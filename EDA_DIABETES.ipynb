```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('diabetes.csv')
```

```python
print(data.head())
```

```python
data.describe(include = 'all')
```

```python
print(data.info())
```

```python
print(data.BMI.unique())
```

```python
#calculate mean of a column
print('Mean: ',data.Glucose.mean())
 
#calculate median of a column
print('Median: ',data.Glucose.median())
 
#calculate mode of a column
print('Mode: ',data.Glucose.mode())
 
```

```python
chart_data = ['Pregnancies','Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']
```

```python
for f in chart_data:
    sns.histplot(x=f, data= data)
    plt.show()
    plt.close()
```

```python
plt.bar("BloodPressure", "Age", data = data, color = "red")
plt.xlabel("BloodPressure")
plt.ylabel("Age")
plt.title("BloodPressure vs Age")
plt.show()
```

```python
for f in chart_data:
    sns.barplot(x=f, data=data)
    plt.show()
    plt.close()
```

```python
age_counts= data.Age.value_counts()
print(age_counts)
```

```python
preg_counts= data.Pregnancies.value_counts()
print(preg_counts)
```

```python
Insulin_counts= data.Insulin.value_counts()
print(Insulin_counts)
```

```python
bmi_counts= data.BMI.value_counts()
print(bmi_counts)
```

```python
Insulin_props = data.Insulin.value_counts()/len(data.Insulin)
print(Insulin_props)
```

```python
BP_props = data.BloodPressure.value_counts()/len(data.BloodPressure)
print(BP_props)
```

```python
for f in chart_data:
    sns.countplot(x=f, data= data)
    plt.show()
    plt.close()
```
